```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img=cv2.imread('/gdrive/My Drive/Colab Notebooks/Doraemonsky.jpg')
cv2_imshow(img)
print(img.shape)
Blueimage=img[:,:,0]
Greenimage=img[:,:,1]
Redimage=img[:,:,2]
print('Blue image :' )
cv2_imshow(Blueimage)
print('Green image :' )
cv2_imshow(Greenimage)
print('Red image :' )
cv2_imshow(Redimage)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img=cv2.imread('/gdrive/My Drive/Colab Notebooks/aiger.jpg')
cv2_imshow(img)

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray.shape)
cv2_imshow(gray)

ret,thresh1=cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
print(thresh1.shape)
cv2_imshow(thresh1)
ret,thresh1=cv2.threshold(gray,150,255,cv2.THRESH_BINARY_INV)
print(thresh1.shape)
cv2_imshow(thresh1)
```
